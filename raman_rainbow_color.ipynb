```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

file_name = 'raman'
data = pd.read_excel(f'{file_name}.xlsx')
data.head()
```

```python
fig, ax = plt.subplots(figsize= (10.7, 8.2), dpi=150)
label = ['FTO_blank', 'FTO_RT', 'FTO_60', 'FTO_200', 'FTO_200_air', 'FTO_300_air']
c = 0.1

for i in range(0, len(label)*2 - 1, 2):
    c +=0.25
    if i == 10:
        ax.plot(np.array(data.iloc[:, i]),
            np.array(data.iloc[:, i+1])+(0.35*i),
            linewidth = 2,
            label = label[int(i/2)])
    else:
        ax.plot(np.array(data.iloc[:, i]),
                np.array(data.iloc[:, i+1])+(0.4*i),
                linewidth = 1,
                label = label[int(i/2)])

print(len(ax.lines))    
colormap = plt.cm.gist_ncar  
colors = [colormap(i) for i in np.linspace(0, 1, len(ax.lines)/2.7 + 5)]  #change here

for i,j in enumerate(ax.lines):
    j.set_color(colors[i])
    
ax.legend(loc = 'upper right', fontsize = 14)
plt.ylabel('Intensity', fontsize = 30, weight = 'bold')
plt.xlabel("Raman shift (cm$^{-1}$)", fontsize = 30, weight = 'bold')

#plt.text(0.03, 1100, file_name, fontsize = 22, color = 'red')

ax.xaxis.set_tick_params(labelsize=26)
ax.yaxis.set_ticklabels([])
plt.minorticks_on()
ax.set_ylim(0.1, 6)
ax.set_xlim(130, 1100)

ax.spines["bottom"].set_linewidth(2)
ax.spines["top"].set_linewidth(2)
ax.spines["right"].set_linewidth(2)
ax.spines["left"].set_linewidth(2)

#plt.savefig(f'CV_{file_name}.tif', bbox_inches = 'tight')

plt.show()
```
